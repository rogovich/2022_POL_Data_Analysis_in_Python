```python
text = ': %{x}<br> : %{y}'

trace0 = go.Scatter(x=forest.year, 
                    y=forest.hare,
                    marker=dict(color='#fc03f8'),
                    name='',
                    hovertemplate=text)

trace1 = go.Scatter(x=forest.year, 
                    y=forest.lynx,
                    marker={'color':'orange'},
                    name='',
                    hovertemplate=text)

trace2 = go.Scatter(x=forest.year, 
                    y=forest.carrot,
                    marker={'color':'grey'}, 
                    name='',
                    hovertemplate=text)

layout = go.Layout(title = ', , ',
                   xaxis = dict(title=''),
                   yaxis = dict(title=''))

fig = go.Figure([trace0, trace1, trace2], layout)
fig.show()
```
